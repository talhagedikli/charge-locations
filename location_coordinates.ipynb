```python
import pandas as pd
import numpy as np

data = pd.read_csv("/content/drive/Othercomputers/MacBook Air'im/Documents/Data/ilce nufus.csv")
data.head()
```

```python
from geopy.geocoders import Nominatim

def get_coordinates(neighborhood_name, district, city="İstanbul"):
    geolocator = Nominatim(user_agent="my_geocoder")

    # Combine neighborhood name and city for better accuracy
    location = geolocator.geocode(f"{neighborhood_name}, {district}, {city}", language="tr")

    if location:
        latitude, longitude = location.latitude, location.longitude
        return latitude, longitude
    else:
        print(f"Coordinates not found for {neighborhood_name}, {city}")
        return None

# Example usage:
neighborhoods = [("Beşiktaş", "Levent")]

for neighborhood, district in neighborhoods:
    coordinates = get_coordinates(neighborhood, district)
    if coordinates:
        print(f"{neighborhood}, {district}: Latitude {coordinates[0]}, Longitude {coordinates[1]}")
```

```python
# @title
for index, row in data.iterrows():
  neighborhood = row["Mahalle"]
  district = row["Ilce"]
  coords = get_coordinates(neighborhood, district)

  if coords == None:
    coords = get_coordinates("", district)
    print(f"{neighborhood}, {district}: Latitude {coords[0]}, Longitude {coords[1]}")


  data.loc[index, "latitude"], data.loc[index, "longitude"] = coords[0], coords[1]


```
